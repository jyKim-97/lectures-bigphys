```
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import time

position_x = float(input("입자의 x 위치는?"))
position_y = float(input("입자의 y 위치는?"))

memory = np.array([[position_x, position_y]])

para = 0.005

gradient_x = 0
gradient_y = 0

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
 
ax.set_xlim([position_x, position_y])
ax.set_ylim([position_x, position_y])

while (position_x,position_y) != (0,0):
  plt.cla()
  gradient_x = gradient_x -2*position_x
  gradient_y = gradient_y -2*position_y
  position_x, position_y = position_x + (para*gradient_x), position_y + (para*gradient_y)
  print(position_x, position_y )
  plt.scatter(position_x, position_y)
  plt.show()
  time.sleep(1)

```
