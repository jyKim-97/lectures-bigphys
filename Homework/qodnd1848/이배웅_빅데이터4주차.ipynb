```
#x,y,z 위치 랜덤 생성 정확도 확인 
iter = int(input("몇 개의 예를 보고 싶으신가요?"))

radius = 5

def create_dot():
    x = ((random.randint(0,1)*2)-1)*random.triangular(0, radius,radius)
    y = ((random.randint(0,1)*2)-1)*random.triangular(0, radius,radius)
    z = ((random.randint(0,1)*2)-1)*random.triangular(0, radius,radius)

    return  x, y, z

x, y, z = create_dot()

#force_x = 0
x_total = 0
x_list = np.array([])

#force_y = 0
y_total = 0
y_list = np.array([])

#force_z = 0
z_total = 0
z_list = np.array([])

count = 0
while count < iter:
  x, y, z = create_dot()
  x_total = x_total + x
  x_list = np.append(x_list, x)
  y_total = y_total + y
  y_list = np.append(y_list, y)
  z_total = z_total + z
  z_list = np.append(z_list, z)
  process = count/iter *100
  if process % 10 == 0:
    print(process)
  count += 1

print(x_total)
print(y_total)
print(z_total)
print("-------------------------------")
plt.plot(x_list, y_list, '.')
plt.plot(x_list, z_list, '.')
plt.plot(y_list, z_list, '.')
plt.show
```

```
#x,y,z에 걸리는 힘 측정
iter = int(input("몇 개의 예를 보고 싶으신가요?"))

#구체의 반지름은 5, 전하밀도의 상수는 1로 두었다.
k_const = 1
radius = 5

#쿨롱 힘 상수(9x10^9N)와 점전하의 전하량역시 편의상 1로 두었다.
def create_dot():
    x = ((random.randint(0,1)*2)-1)*random.triangular(0, radius,radius)
    y = ((random.randint(0,1)*2)-1)*random.triangular(0, radius,radius)
    z = ((random.randint(0,1)*2)-1)*random.triangular(0, radius,radius)
    force_x = 1/((x-10)**2)
    force_y = (y/abs(y))/(y**2)
    force_z = (z/abs(z))/(z**2)
    

    return force_x, force_y, force_z

count = 0

#force_x = 0
force_x_total = 0
force_x_list = np.array([])

#force_y = 0
force_y_total = 0
force_y_list = np.array([])

#force_z = 0
force_z_total = 0
force_z_list = np.array([])

#최종 힘(전하의 개수만큼 normalize 되야한다. 만약 normalize안되면 입자의 개수만큼 정비례하여 힘이 커진다)
total_force = 0
total_force_list = np.array([])

process = 0

while count < iter:
  force_x, force_y, force_z = create_dot()

  force_x_total += force_x
  force_x_list = np.append(force_x_list, force_x_total/(count+1))
  force_y_total += force_y
  force_y_list = np.append(force_y_list, force_y_total/(count+1)) 
  force_z_total += force_z 
  force_z_list = np.append(force_z_list, force_z_total/(count+1))

  total_force = ((((force_x_total)**2) + ((force_x_total)**2) + ((force_x_total)**2))**0.5)/(count+1)
  total_force_list = np.append(total_force_list, total_force)
  
  process = count/iter *100
  if process % 5 == 0:
    print(process,"%")
    print("Force =",total_force)
  count += 1



plt.plot(force_x_list, 'r')
plt.plot(force_y_list, 'b')
plt.plot(force_z_list, 'g')
plt.plot(total_force_list, 'k')
plt.show
```
