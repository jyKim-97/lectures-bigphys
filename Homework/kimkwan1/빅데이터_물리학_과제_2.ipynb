```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```
class NN:                         
  def __init__(self,N):
    self.N=N
    
  def makeNN(self):                      ## NxN 빈행렬 만들기.
    self.P = np.zeros((self.N,self.N))
  
  def BC(self,a,b,c,d):                  ## 행렬의 boundary condition 설정.          
    for i in range(self.N):     
      self.P[self.N-1][i]=b
      self.P[i][self.N-1]=c
      self.P[i][0]=d
      self.P[0][i]=a                
   
  def op1(self,k):                       ## 평균값을 구하는 식을 k번 실행한다.
    for i in range(k):
      for i in range(1,self.N-1):
        for j in range(1,self.N-1):
          self.P[i][j]=(self.P[i-1][j]+self.P[i+1][j]+self.P[i][j-1]+self.P[i][j+1])/4
    return self.P      
  
  def op2(self,k):                       ## (b) (5,5) 에 입자가 있는경우.
    for i in range(k):
      for i in range(1,self.N-1):
        for j in range(1,self.N-1):
          self.P[i][j]=(self.P[i-1][j]+self.P[i+1][j]+self.P[i][j-1]+self.P[i][j+1])/4
          self.P[10][10]=10
    return self.P      
  
                                                    
  def visual(self):                     ## visualization
    fig = plt.figure()
    ax = fig.gca(projection='3d')
    for i in range(0,self.N-1):
      for j in range(0,self.N-1):
        ax.scatter(i,j,self.P[i][j],c='r')
    return plt.show()
```


(a) 경계조건 (0<=x<=10 , 0<=y<=10)
---
  x=0    ->    v=10                                                                                                
  x=10  ->    v=0                                               
  y=0    ->    v=0             
  y=10  ->    v=0




```
a=NN(20)   
```

```
a.makeNN()
```

```
a.BC(10,0,0,0)     ## V대신 임의의 값 10으로 실행
```

```
a.op1(1000)
```

```
a.P[1][3]
```

```
a.visual()
```

(c) 경계조건 (0<=x<=1 , 0<=y<=1)
---
  x=0    ->    v=0                                                                                                
  x=1    ->    v=0                                               
  y=0    ->    v=sin(πx)             
  y=1    ->    v=sin(πx)


```
c=NN(21)      ## N x N
```

```
c.makeNN()
```

```
for i in range(21):              ## range N     
    c.P[i][0]=np.sin(np.pi*1/20*i)           
    c.P[i][20]=np.sin(np.pi*1/20*i)        ## P[i][N-1] = np.sin(np.pi*1/N*i)
    c.P[20][i]=0
    c.P[0][i]=0       
    
```

```
c.op1(1000)
```

```
c.visual()
```
