```
#빅데이터 물리학 과제 2

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import tqdm
```

```
image_size = 100 # 이미지 크기
Voltage = 10.0 # 전압
grid = np.zeros([image_size + 1, image_size + 1])
```

```
arr_1 = np.zeros_like(grid, float)
arr_2 = np.zeros_like(grid, float)
arr_1[0, :] = Voltage
arr_2[0, :] = Voltage
for a in tqdm.tqdm(range(3000)):
    for i in range(1, image_size):
        for j in range(1, image_size):
            arr_2[i, j] = (arr_1[i - 1, j] + arr_1[i + 1, j] + arr_1[i, j - 1] + arr_1[i, j + 1]) / 4
    delta = np.max(abs(arr_1 - arr_2)) #오차?
    arr_1,arr_2 = arr_2,arr_1
```

```
plt.imshow(arr_1)
plt.savefig("Bigdata arr_1.png")
arr_1_table = pd.DataFrame(arr_1)
arr_1_table.to_csv('Data.csv')
```
