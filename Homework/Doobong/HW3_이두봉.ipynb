```
import numpy as np
import random
import matplotlib.pyplot as plt
from math import *
```

```
R = 5 # 구의 반지름
nExp = 10000 # 입자 갯수
x_list, y_list, z_list = [], [], []
a_list = []
force_x_list, force_y_list, force_z_list =[], [], []
```

```
for i in range(nExp):
  x = random.uniform(-2, 2)
  y = random.uniform(-2, 2)
  z = random.uniform(-2, 2)
  r = np.sqrt(x**2 + y**2 + z**2) 
  
  

  if (x**2 + y**2 + z**2 <= 4):
      x_list.append(x)
      y_list.append(y)
      z_list.append(z)
      a = 1/r**2 * np.array([x, y, z])
      a_list.append(a)
  
  else:
      pass
```

```
plt.plot(x_list,y_list,'.')
plt.plot(x_list,z_list,'.')
plt.show()
```

```
for i in range(len(x_list)):
  d = np.sqrt((x_list[i]-10)**2+y_list[i]**2+z_list[i]**2)
  unit_vector =  [(10 - x_list[i])/d, -y_list[i]/d, -z_list[i]/d]
  force_s = 100 / d**2
  force_v_x = unit_vector[0]*force_s
  force_v_y = unit_vector[1]*force_s
  force_v_z = unit_vector[2]*force_s
  force_x_list.append(force_v_x)
  force_y_list.append(force_v_y)
  force_z_list.append(force_v_z)
```

```
plt.hist(force_x_list, bins=1000)
plt.show()
plt.hist(force_y_list, bins=1000)
plt.show()
plt.hist(force_z_list, bins=1000)
plt.show()
```

```
force_mean = [0,0,0]
for i in range(len(force_x_list)):
  force_mean[0] += force_x_list[i]
  force_mean[1] += force_y_list[i]
  force_mean[2] += force_z_list[i]

for i in range(3):
  force_mean[i] /= len(force_x_list)

print(force_mean)
```
